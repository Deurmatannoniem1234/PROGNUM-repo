```python
import matplotlib.pyplot as plt
import numpy as np
from math import cos, sin

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

def Rotate(x, y, a):
    rad=np.radians(a)
    xmid = np.average(x[range(4)])
    ymid = np.average(y[range(4)])
    x0 = x - xmid
    y0 = y - ymid
    xa = (x0)*cos(rad) - (y0)*sin(rad) + xmid
    ya = (x0)*sin(rad) + (y0)*cos(rad) + ymid
    return(xa, ya)

xa, ya = Rotate(x, y, 30)


plt.plot(x,y)
plt.plot(xa,ya)
plt.xlim(0,4)
plt.ylim(0,4)
plt.gca().set_aspect('equal')
plt.show
```

```python
import numpy as np

def funcfilter(func, seq):
    arr = np.asarray(seq)[np.asarray(seq) >0]
    result = func(arr)
    return result

sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)

sq = 100
result = funcfilter(np.log10, sq)
print("Result log10:", result)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

x_axis = np.linspace(-6, 8, 1000)



# 1D Histogram 
plt.figure(figsize=(10, 6))
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.hist(X, bins=30, alpha=0.5, density=True, label="X", color='blue', edgecolor='black')
plt.hist(Y, bins=30, alpha=0.5, density=True, label="Y", color='red', edgecolor='black')

plt.plot(x_axis, norm.pdf(x_axis, 0, 1), color='black', linestyle=':', linewidth=4, label="X(PDF)")
plt.plot(x_axis, norm.pdf(x_axis, 2, 1), color='black', linestyle='--', linewidth=4, label="Y(PDF)")

plt.xticks(np.arange(-6, 9, 1))
plt.xlim(-6,8)
plt.ylim(0,0.42)
plt.title("1D Histogram + PDF")
plt.xlabel("Values")
plt.ylabel("Probability Density")

plt.annotate(r'$\mathcal{N}(0, 1)$',xy=(-1.8, 0.15), xytext=(-5.5, 0.145),
             arrowprops=dict(facecolor='blue', shrink=0.05), color='blue', fontsize=15)
plt.annotate(r'$\mathcal{N}(2,1)$',xy=(4, 0.15), xytext=(6, 0.145),
             arrowprops=dict(facecolor='r', shrink=0.05), color='r', fontsize=15)
plt.legend()
plt.show()




#2D Histogram
bins = 30
range_lim = [[-17, 17],[-17,17]]

H, xedges, yedges = np.histogram2d(X, Y, bins=100, range = range_lim)

# Bin centers (needed for contour)
xcenters = (xedges[:-1] + xedges[1:]) / 2
ycenters = (yedges[:-1] + yedges[1:]) / 2

# Meshgrid for contour
Xgrid, Ygrid = np.meshgrid(xcenters, ycenters)

# 1. Flatten and sort the histogram values in descending order
H_flat = H.flatten()
H_sort = np.sort(H_flat)[::-1]

# 2. Calculate the cumulative sum of the sorted bins
cumulative_sum = np.cumsum(H_sort)
cumulative_prob = cumulative_sum / cumulative_sum[-1]

target_probs = [0.68, 0.95, 0.997]
levels = []

for p in target_probs:
    # Find the threshold bin value
    idx = np.where(cumulative_prob >= p)[0][0]
    levels.append(H_sort[idx])

levels = sorted(levels)

plt.figure(figsize=(8, 7))
plt.hist2d(X,Y, bins = bins, cmap = "Blues")
plt.colorbar(label='')
plt.grid(True, linestyle='--', color='black', alpha=0.7)


contours = plt.contour(
    Xgrid, Ygrid, H.T,
    levels=levels,
    colors="red",
    linewidths=2
)

fmt = {
    levels[2]: r'$1\sigma$',
    levels[1]: r'$2\sigma$',
    levels[0]: r'$3\sigma$'
}

plt.clabel(contours, contours.levels, inline=True, fmt=fmt, fontsize=12, colors='red')

plt.title(f"2D Histogram + Contour")


plt.show()
```
