```python
from scipy.constants import G
import numpy as np
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
EV = []
for i in range(len(masses)):
    EV.append(round(((2*G*masses[i]/radii[i])**0.5),3))
    print(f"{EV[i]:6.2f}")

NandEV= dict(zip(names,EV))
print(f"{NandEV}")
```

```python
from math import pi
from IPython.display import display, Math
def Cylinder(r,h):
    Area = 2*pi*r**2 + 2*pi*r*h
    Volume = pi*r**2*h
    return(Area, Volume)
a = Cylinder(2,3)
ratio = rf"\text{{The ratio between the Area and Volume is = }}{a[0]/a[1]:.2f}"
display(Math(ratio))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def evalf(Fx, x):
    return(eval(Fx))

x = np.array(range(-5,6))
xnp = np.arange(-5,6,0.1)
y = evalf("(np.exp(x)+np.exp(-x))/2", x)
ynp = evalf("(np.exp(x)+np.exp(-x))/2", xnp)

plt.plot(x,y, label="catenary line x")
plt.plot(xnp,ynp, label="catenary line xnp")
plt.title("cosh(x)")
plt.grid(True)
plt.xlabel("x", fontsize = 12)
plt.ylabel("cosh(x)", fontsize = 12)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(fontsize=8)
plt.show
```
