```python
# Sirius data
apparentMagnitude = float(input("What is the Apparent Magnitude?"))
absoluteMagnitude = float(input("What is the Absolute Magnitude?"))

# The distance is related to the magnitudes as m-M=5.Log(d/10)
Parsec = 3.26164

m = apparentMagnitude
M = absoluteMagnitude

d = 10.0 * pow( 10.0, (m-M)/5.0 ) * 3.26164
print(f"The distance to this star is {d} pc and thus {d*Parsec} ly")
```
